**縮排 (indentation)**
- Python 區分程式區塊 (block) 的方式，所謂程式區塊是程式中具有特定功能之處，程式區塊依據條件真或假的結果個別執行。所謂縮排也就是在程式碼開頭的地方多出一些空格數。


if

```python
a=input('輸入一個大於5的數字')
a=int(a)
if (a>5):#if
    print('你好棒')
```

If….else

```python
a=input('輸入一個大於5的數字     ')
a=int(a)
if (a>5):
    print('你好棒')
else:
    print('你不棒')
```

If….elif…else

```python
a=input('輸入一個大於5的數字     ')
a=int(a)
if(a>5):
    print('你好棒')
elif(a==5):
    print('再..大..一..點..點')
else:
    print('你不棒')
```

巢狀判斷式

```python
a=input('輸入一個大於5的數字     ')
a=int(a)
b=input('輸入一個大於6的數字     ')
b=int(b)
if(a>5):
    if(b>6):
        print('你好棒')
    else:
        print('你好壞')    
else:
    print('你不棒')
```
